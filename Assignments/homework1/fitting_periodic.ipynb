```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
def periodic(x, a, b, c, d, e):
    return a + b * x + c * np.sin(e * x) + d * np.cos(e * x)
```

```python
a0 = 0
a1 = 1
a2 = 1
a3 = 1
a4 = 0
t = np.linspace(20, 35, 100)
y = periodic(t, a0, a1, a2, a3, a4) + np.random.normal(0, 0.5, 100)

# dataset is t, y
```

```python
popt, pcov = curve_fit(periodic, t, y)
popt
```

```python
#are the fit parameters close to the original parameters?
np.array([a0, a1, a2, a3, a4]) - popt
```

it's all pretty close except for a1 which is about 0.5 off of being the correct value

```python
plt.plot(t, y, 'b-', label='data')
plt.plot(t, periodic(t, *popt), 'r-', label='fit')
plt.legend()
```

```python
def periodic2(x, a, b, c, d, e):
    return a + b*x + c*np.sin(d*x + e)
```

```python
popt2, pcov2 = curve_fit(periodic2, t, y)
```

```python
plt.plot(t, y, 'k-', label='data')
plt.plot(t, periodic(t, *popt), 'r-', label='fit_func1', alpha=0.5)
plt.plot(t, periodic2(t, *popt2), 'b-', label='fit_func2', alpha=0.5)
plt.legend()
# set the alpha to 0.5, so if these points intersect, it'll be purple
```
